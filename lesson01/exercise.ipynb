```python
def exercise_4_security_preprocessor(raw_data):
    """
    å®‰å…¨æ•°æ®é¢„å¤„ç†å™¨
    
    å‚æ•°:
        raw_data (list): åŸå§‹å®‰å…¨æ•°æ®ï¼Œæ¯ä¸ªæ¡ç›®åŒ…å«: ip, timestamp, bytes, duration, protocol
    
    è¿”å›:
        dict: é¢„å¤„ç†åçš„æ•°æ®
    """
    processed_data = {
        "cleaned_data": [],           # æ¸…æ´—åçš„æ•°æ®
        "feature_matrix": None,       # ç‰¹å¾çŸ©é˜µï¼ˆNumPyæ•°ç»„ï¼‰
        "feature_names": [],          # ç‰¹å¾åç§°
        "summary": {}                 # å¤„ç†æ‘˜è¦
    }
    
    # TODO: åœ¨è¿™é‡Œå®ç°æ‚¨çš„ä»£ç 
    # æ‚¨çš„ä»£ç å¼€å§‹ ==================
    
    
    
    # æ‚¨çš„ä»£ç ç»“æŸ ==================
    
    return processed_data

# è¾…åŠ©å‡½æ•°ï¼šæ£€æŸ¥IPåœ°å€æ ¼å¼æ˜¯å¦æœ‰æ•ˆ
def is_valid_ip(ip):
    """
    æ£€æŸ¥IPåœ°å€æ ¼å¼æ˜¯å¦æœ‰æ•ˆï¼ˆç®€åŒ–ç‰ˆæœ¬ï¼‰
    """
    if not ip or not isinstance(ip, str):
        return False
    
    parts = ip.split('.')
    if len(parts) != 4:
        return False
        
    try:
        for part in parts:
            num = int(part)
            if not (0 <= num <= 255):
                return False
        return True
    except ValueError:
        return False

# è¾…åŠ©å‡½æ•°ï¼šæ ‡å‡†åŒ–æ•°ç»„
def normalize_array(arr):
    """
    æ ‡å‡†åŒ–æ•°ç»„ï¼ˆZ-scoreæ ‡å‡†åŒ–ï¼‰
    """
    mean = np.mean(arr)
    std = np.std(arr)
    if std == 0:
        return np.zeros_like(arr)
    return (arr - mean) / std
```
